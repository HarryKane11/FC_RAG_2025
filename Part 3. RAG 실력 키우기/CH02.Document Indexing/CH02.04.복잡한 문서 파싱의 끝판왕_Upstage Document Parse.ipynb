```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import os
from langchain_upstage import UpstageDocumentParseLoader
 
file_path = "data/국가별 공공부문 AI 도입 및 활용 전략.pdf"
loader = UpstageDocumentParseLoader(file_path, ocr="auto", split="page")

pages = loader.load_and_split()
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(pages[8].page_content, 'html.parser')

import pandas as pd

data = []

for tag in soup.find_all():
    data.append({
        'tag': tag.name,
        'id': tag.get('id'),
        'text': tag.get_text(strip=True),
        'style': tag.get('style')
    })

df = pd.DataFrame(data)
display(df)
```
